```python
# Import needed dependencies
import requests
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import re
import time
from datetime import date
from bs4 import BeautifulSoup, Comment
from unidecode import unidecode
import urllib.request
import requests
import shutil

pd.set_option('display.max_columns', None)
```

```python
player_stats_df = pd.read_csv('NBA_all_player_stats.csv')
player_stats_df.drop('Unnamed: 0', axis = 1, inplace = True)
player_stats_df
```

```python
gamescore_leaders_df = player_stats_df.sort_values('GS',ascending = False).head(10)
```

```python
gamescore_leaders_df
```

```python
# https://www.basketball-reference.com/players/b/brownja02.html

top_gs_players_list = gamescore_leaders_df['Name'].to_list()
unique_top_gs_players_list = list(set(top_gs_players_list))
```

```python
url_names = []
for item in unique_top_gs_players_list:
    split_name = item.split()
    url_part1 = re.findall(r"^.{1}", split_name[1])
    url_part2 = re.findall(r"^.{2}", split_name[0])
    url_part3 = re.findall(r"^.{5}", split_name[1])
    url_part3 = url_part3[0].split()
    
    url_part1 = unidecode(url_part1[0]).lower()
    url_part2 = unidecode(url_part2[0]).lower()
    url_part3 = unidecode(url_part3[0]).lower()
    
    url_names.append([url_part1,url_part2,url_part3])
    
```

```python
unique_top_gs_players_list
```

```python
url_list = []
for i in range(len(unique_top_gs_players_list)):
    time.sleep(5)
    
    print(unique_top_gs_players_list[i])
    
    url_part4_int = 1
    url_part4 = '0'+str(url_part4_int)
    url = f'https://www.basketball-reference.com/players/{url_names[i][0]}/{url_names[i][2]}{url_names[i][1]}{url_part4}.html'
        
    soup = BeautifulSoup(requests.get(url).content, 'html.parser')
    
    if unique_top_gs_players_list[i] in soup.title.get_text():
        url_list.append(url)
        continue
    else:
        while unique_top_gs_players_list[i] not in soup.title.get_text():
            url_part4_int = url_part4_int + 1
            url_part4 = '0'+str(url_part4_int)
                        
            url = f'https://www.basketball-reference.com/players/{url_names[i][0]}/{url_names[i][2]}{url_names[i][1]}{url_part4}.html'
            soup = BeautifulSoup(requests.get(url).content, 'html.parser')

            if unique_top_gs_players_list[i] in soup.title.get_text():
                url_list.append(url)
    
     
    
    
    
```

```python
for i in range(len(url_list)):
    time.sleep(5)
  
    html_page = urllib.request.urlopen(url_list[i])
    soup = BeautifulSoup(html_page, features="lxml")
    for img in soup.findAll('img'):
        if unique_top_gs_players_list[i] in str(img):
            assa=(img.get('src'))
            urllib.request.urlretrieve(assa, f"/Users/michaelbinger/Desktop/Sport_Logos/NBA_photos/Game_score_leaders/{unique_top_gs_players_list[i]}.png")

    
    
```
